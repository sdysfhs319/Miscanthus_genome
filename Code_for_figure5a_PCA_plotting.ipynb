```r
library(ggplot2)
data <-read.table(file="vcf_for_analysis.snp.pca.evec_for_plotting",header=T, sep="\t")
data
```

```r
p2 <- ggplot(data = data, aes(x = PC1, y = PC2)) + geom_point(aes(color=Group), size=6,alpha=0.6) + theme_classic()+ labs(x="PC1 (7.66%)", y= "PC2 (5.42%)") +
theme(axis.text.x = element_text(size = 14, color="black"), axis.text.y = element_text(size = 14, color="black"),
                       axis.title.x = element_text(size = 16, face = "bold"), axis.title.y = element_text(size = 16, face = "bold"),legend.position=c(0.8,0.85), 
                       legend.text = element_text(size = 16, face="italic"),legend.title=element_blank() )
p2
```

```r
ggsave("Reseq_PCA_plotting_new.png", p2, height=5.9, width=6.67, dpi=600)
```
