```r
library(ggplot2)
library(forcats)
library(ggthemes)
#install.packages(c("forcats", "ggthemes"))
```

```r
df1 = read.table("K_CV", sep="\t", header=T)
df1
```

```r
kv <- ggplot(data=df1, aes(x=K, y=Cross.validation.error)) + geom_point() + geom_line() + theme_classic() +
theme(axis.text.x = element_text(size = 14, color="black"), axis.text.y = element_text(size = 14, color="black") ,
                       axis.title.x = element_text(size = 16), axis.title.y = element_text(size = 16))
(kv_final <- kv + scale_x_discrete(breaks=seq(1,7),limits=c("1","2","3","4","5","6","7")) + ylab("Cross validation error"))
### breaks控制在哪个位置放坐标轴标签，limits控制放标签的名称和顺序
```

```r
ggsave("CV.png", kv_final, dpi=600)
```

```r
df3 <- read.table("admixture_k_3", header=T, sep="\t")
df3
```

```r
k3plot <-
  ggplot(df3, aes(factor(sample_ID), value, fill = factor(variable))) +
  geom_col(color="gray", size=0.2) +
  facet_grid(~factor(Species, levels=c("M. floridulus", "M. sacchariflorus", "M. sinensis", "M. lutarioriparius", "Mg", "Hybrid")), scales = "free_x", switch = "x", space = "free") +
  theme_classic() +  scale_y_continuous(expand = c(0, 0)) +
  scale_x_discrete(expand = expand_scale(add = 0.5)) + theme(
    panel.spacing.x = unit(0.5, "lines"), strip.text.x = element_text(size = 20, colour = "black", face="bold.italic"),
    axis.text.x = element_text(size = 18, color = "black", vjust = 1, hjust = 1, angle = 45),
     axis.text.y = element_text(size = 18, color = "black", face = "bold"), axis.title = element_blank(), 
  ) +  scale_fill_gdocs(guide = FALSE)
k3plot
ggsave("admixture_3.png", k3plot, height = 6, width=24, dpi=600)
```

```r
ggsave("admixture_3.pdf", k3plot, height = 6, width=24)
```
