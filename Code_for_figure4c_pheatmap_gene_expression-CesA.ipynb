```r
library(ggplot2)
library(pheatmap)
library(reshape2)
```

```r
df1 <- read.table("MFLOR_CesA.txt", header = F,check.names = F)
df1
```

```r
colnames(df1) <- c("Gene_name", "root_1", "root_2", "root_3", "stem_1", "stem_2", "stem_3", "leaf_1", "leaf_2", "leaf_3", "inflorescence_1", "inflorescence_2", "inflorescence_3","Gene_family") 
```

```r
df2 <- df1[apply(df1[2:13], 1, function(x) sd(x)!=0),]
df2
```

```r
df4 <- df2[apply(df2[2:13], 1, function(x) mean(x) >= 1),]
df4
```

```r
df3 <- cbind(df4[1], log2(df4[2:13] + 0.5), df4[14]) 
df3
```

```r
gene_group <- read.table("MFLOR_CesA_test2", header = T)
gene_group
```

```r
gene_group$Gene_family <- factor(gene_group$Gene_family, levels = c("CesA1","CesA2", "CesA3and5", "CesA4","CesA6", "CesA7", "CesA8", "CesA9","CesAX",
                                                                "CslD", "CslE", "CslF", "CslH", "CslJ"))
```

```r
gene_group
```

```r
df3_ordered <- df3[rownames(gene_group), ]
df3_ordered
```

```r
df3_ordered_noNA <- na.omit(df3_ordered)
df3_ordered_noNA
```

```r
df4_ordered <- df4[rownames(gene_group), ]
df4_ordered_noNA <- na.omit(df4_ordered)
```

```r
col <- colorRampPalette(c("lightblue", "yellow", "orange", "red"),bias=3)(14)
col
```

```r
ann_colors = list( Gene_family = c(CesA1='#FF6347', CesA2='#F08080', CesA3and5='#33ffff', CesA4='#EED5D2', CesA6='#00CD66', CesA7='#FF83FA',
                                   CesA8='#0099cc', CesA9='#BF3EFF',CesAX='#CDAD00',  CslD='#68228B', CslE='#66CD00', CslF='orange',
                                  CslH='#00E5EE', CslJ='darkblue'))
```

```r
p<-pheatmap(df3_ordered_noNA[2:13],scale = "none", annotation_colors = ann_colors,show_rownames = F,cluster_rows =F,clustering_method = "average", display_numbers=F,
            annotation_row=gene_group, cluster_cols = F,anglecol = 90, fontsize=14,border_color="white", annotation_legend=T)


p
```
